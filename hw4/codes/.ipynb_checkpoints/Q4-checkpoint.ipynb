```python
import torch
from torchvision import datasets, transforms
import random
import matplotlib.pyplot as plt
from torch.utils.data import random_split
import numpy as np


#Loading datasets

transform = transforms.ToTensor()


mnist_train = datasets.MNIST(root='./hw4/codes', train=True, download=True, transform=transform)

mnist_test = datasets.MNIST(root='./hw4/codes', train=False, download=True, transform=transform)
```

```python

mnist_train_data = np.array(mnist_train.data)
mnist_train_labels = np.array(mnist_train.targets)


mnist_test_data = np.array(mnist_test.data)
mnist_test_labels = np.array(mnist_test.targets)


print(mnist_train_data.shape)
print(mnist_train_labels.shape)
```

```python
def flatten_images(images):
    flattened_images = images.reshape(images.shape[0], -1)
    return flattened_images

mnist_train_1D = flatten_images(mnist_train_data)
mnist_test_1D=flatten_images(mnist_test_data)
```

```python
def normalize_to_01(flattened_images):
    return flattened_images / 255.0


mnist_tr=normalize_to_01(mnist_train_1D)
mnist_ts=normalize_to_01(mnist_test_1D)
```

```python
import numpy as np

def relu(x):
    return np.maximum(0, x)

def d_relu(x):
    return (x > 0).astype(float)


def softmax(z):
    exp_z = np.exp(z - np.max(z, axis=1, keepdims=True))
    return exp_z / np.sum(exp_z, axis=1, keepdims=True)

def grad_softmax_cross_entropy_with_logits(logits, y_true):
    m = logits.shape[0]
    softmax_vals = softmax(logits)
    softmax_vals[np.arange(m), y_true] -= 1
    return softmax_vals/m

def cross_entropy(predictions, targets, epsilon=1e-12):
    predictions = np.clip(predictions, epsilon, 1. - epsilon)
    N = predictions.shape[0]
    return -np.sum(targets*np.log(predictions))/N

def delta_cross_entropy(pred, real):
    n_samples = real.shape[0]
    grad = pred.copy()
    grad[np.arange(n_samples), real.argmax(axis=1)] -= 1
    grad = grad/n_samples
    return grad

def one_hot_encoding(labels, num_classes):
    return np.eye(num_classes)[labels]
```

```python

#####Q4.2 using numpy
class MLP:
    def __init__(self, input_size, output_size, hidden_layers=[64, 64]):
        self.weights = []
        self.biases = []
        
        layers = [input_size] + hidden_layers + [output_size]
        for i in range(len(layers) - 1):
            self.weights.append(np.random.randn(layers[i], layers[i+1]) * np.sqrt(2./layers[i]))
            self.biases.append(np.zeros((1, layers[i+1])))

    def forward(self, x):
        self.a = [x]
        self.z = []
        for i, (w, b) in enumerate(zip(self.weights, self.biases)):
            z = np.dot(self.a[-1], w) + b
            self.z.append(z)
            if i == len(self.weights) - 1:
                a = softmax(z)
            else:
                a = relu(z)
            self.a.append(a)
        return self.a[-1]
####### Q4.1
    def backpropagation(self, x, y, learning_rate):
        m = x.shape[0]
        dz = grad_softmax_cross_entropy_with_logits(self.z[-1], np.argmax(y, axis=1))
        dw = np.dot(self.a[-2].T, dz)
        db = np.sum(dz, axis=0, keepdims=True)

        grad_clip = 5.0
        for grad in [dw, db]:
            np.clip(grad, -grad_clip, grad_clip, out=grad)

        self.weights[-1] -= learning_rate * dw
        self.biases[-1] -= learning_rate * db

        for i in range(len(self.weights)-2, -1, -1):
            dz = np.dot(dz, self.weights[i+1].T) * d_relu(self.z[i])
            dw = np.dot(self.a[i].T, dz)
            db = np.sum(dz, axis=0, keepdims=True)

            for grad in [dw, db]:
                np.clip(grad, -grad_clip, grad_clip, out=grad)

            self.weights[i] -= learning_rate * dw
            self.biases[i] -= learning_rate * db

    def evaluate(self, x_val, y_val):
        
        predictions = self.predict(x_val)
        true_labels = np.argmax(y_val, axis=1)
        accuracy = np.mean(predictions == true_labels)
        return accuracy

    def fit(self, x, y, learning_rate, epochs, x_val=None, y_val=None):
        
        self.losses = []
        self.val_accuracies = []
        
        for epoch in range(epochs):
            output = self.forward(x)
            self.backpropagation(x, y, learning_rate)
            loss = cross_entropy(output, y)
            self.losses.append(loss)
            
            if x_val is not None and y_val is not None:
                val_accuracy = self.evaluate(x_val, y_val)
                self.val_accuracies.append(val_accuracy)
                if epoch % 10 == 0:
                    print(f"Epoch {epoch}, Loss: {loss:.3f}")
                    print(f"Validation Accuracy at Epoch {epoch}: {val_accuracy*100:.2f}%")
        
       
        self.plot_training_curves()
    def predict(self, x):
        return np.argmax(self.forward(x), axis=1)
    
    def plot_training_curves(self):
        
        plt.figure(figsize=(12, 4))
        plt.subplot(1, 2, 1)
        plt.plot(self.losses, label='Training Loss')
        plt.title('Training Loss Curve')
        plt.xlabel('Epochs')
        plt.ylabel('Loss')
        plt.legend()

        
        plt.subplot(1, 2, 2)
        plt.plot(self.val_accuracies, label='Validation Accuracy')
        plt.title('Validation Accuracy Curve')
        plt.xlabel('Epochs')
        plt.ylabel('Accuracy')
        plt.legend()
        
        plt.tight_layout()
        plt.show()

    

```

```python
import torch
import torch.nn as nn
import torch.optim as optim
from torchvision import datasets, transforms
import matplotlib.pyplot as plt
#### Q4.3
# Loading datasets
transform = transforms.ToTensor()

mnist_train = datasets.MNIST(root='./data', train=True, download=True, transform=transform)
mnist_test = datasets.MNIST(root='./data', train=False, download=True, transform=transform)

train_loader = torch.utils.data.DataLoader(mnist_train, batch_size=32, shuffle=True)
test_loader = torch.utils.data.DataLoader(mnist_test, batch_size=32)

class MLP(nn.Module):
    def __init__(self, input_size, output_size, hidden_layers=[64, 64]):
        super(MLP, self).__init__()
        layers = []
        layers_dims = [input_size] + hidden_layers + [output_size]
        for i in range(len(layers_dims) - 1):
            layers.append(nn.Linear(layers_dims[i], layers_dims[i+1]))
            if i != len(layers_dims) - 2:  # Not the last layer
                layers.append(nn.ReLU())
        self.net = nn.Sequential(*layers)
    
    def forward(self, x):
        return self.net(x)

model = MLP(input_size=784, output_size=10, hidden_layers=[300, 200])
criterion = nn.CrossEntropyLoss()
optimizer = optim.Adam(model.parameters(), lr=0.001)

train_losses = []
val_accuracies = []

for epoch in range(500):
    # Training
    model.train()
    running_loss = 0.0
    for images, labels in train_loader:
        images = images.view(-1, 784)
        optimizer.zero_grad()
        outputs = model(images)
        loss = criterion(outputs, labels)
        loss.backward()
        optimizer.step()
        running_loss += loss.item()

    train_losses.append(running_loss/len(train_loader))

    # Validation
    model.eval()
    correct = 0
    total = 0
    with torch.no_grad():
        for images, labels in test_loader:
            images = images.view(-1, 784)
            outputs = model(images)
            _, predicted = outputs.max(1)
            total += labels.size(0)
            correct += (predicted == labels).sum().item()

    val_accuracies.append(correct/total)

    if epoch % 1 == 0:
        print(f"Epoch {epoch}, Loss: {running_loss/len(train_loader):.3f}")
        print(f"Validation Accuracy at Epoch {epoch}: {correct/total*100:.2f}%")

# Plotting
plt.figure(figsize=(12, 4))

plt.subplot(1, 2, 1)
plt.plot(train_losses, label='Training Loss')
plt.title('Training Loss Curve')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend()

plt.subplot(1, 2, 2)
plt.plot(val_accuracies, label='Validation Accuracy')
plt.title('Validation Accuracy Curve')
plt.xlabel('Epochs')
plt.ylabel('Accuracy')
plt.legend()

plt.tight_layout()
plt.show()
```

```python
#####Q4.4


class MLP:
    def __init__(self, input_size, output_size, hidden_layers=[64, 64], init_method='he'):
        self.weights = []
        self.biases = []
        
        layers = [input_size] + hidden_layers + [output_size]
        for i in range(len(layers) - 1):
            if init_method == 'zero':
                self.weights.append(np.zeros((layers[i], layers[i+1])))
            elif init_method == 'uniform_random':
                self.weights.append(np.random.uniform(-1, 1, size=(layers[i], layers[i+1])))
            else:  # Default to He initialization
                self.weights.append(np.random.randn(layers[i], layers[i+1]) * np.sqrt(2./layers[i]))
            
            self.biases.append(np.zeros((1, layers[i+1])))

    def forward(self, x):
        self.a = [x]
        self.z = []
        for i, (w, b) in enumerate(zip(self.weights, self.biases)):
            z = np.dot(self.a[-1], w) + b
            self.z.append(z)
            if i == len(self.weights) - 1:
                a = softmax(z)
            else:
                a = relu(z)
            self.a.append(a)
        return self.a[-1]

    def backpropagation(self, x, y, learning_rate):
        m = x.shape[0]
        dz = grad_softmax_cross_entropy_with_logits(self.z[-1], np.argmax(y, axis=1))
        dw = np.dot(self.a[-2].T, dz)
        db = np.sum(dz, axis=0, keepdims=True)

        grad_clip = 5.0
        for grad in [dw, db]:
            np.clip(grad, -grad_clip, grad_clip, out=grad)

        self.weights[-1] -= learning_rate * dw
        self.biases[-1] -= learning_rate * db

        for i in range(len(self.weights)-2, -1, -1):
            dz = np.dot(dz, self.weights[i+1].T) * d_relu(self.z[i])
            dw = np.dot(self.a[i].T, dz)
            db = np.sum(dz, axis=0, keepdims=True)

            for grad in [dw, db]:
                np.clip(grad, -grad_clip, grad_clip, out=grad)

            self.weights[i] -= learning_rate * dw
            self.biases[i] -= learning_rate * db

    def evaluate(self, x_val, y_val):
        
        predictions = self.predict(x_val)
        true_labels = np.argmax(y_val, axis=1)
        accuracy = np.mean(predictions == true_labels)
        return accuracy

    def fit(self, x, y, learning_rate, epochs, x_val=None, y_val=None):
        
        self.losses = []
        self.val_accuracies = []
        
        for epoch in range(epochs):
            output = self.forward(x)
            self.backpropagation(x, y, learning_rate)
            loss = cross_entropy(output, y)
            self.losses.append(loss)
            
            if x_val is not None and y_val is not None:
                val_accuracy = self.evaluate(x_val, y_val)
                self.val_accuracies.append(val_accuracy)
                if epoch % 10 == 0:
                    print(f"Epoch {epoch}, Loss: {loss:.3f}")
                    print(f"Validation Accuracy at Epoch {epoch}: {val_accuracy*100:.2f}%")
        
       
        self.plot_training_curves()
    def predict(self, x):
        return np.argmax(self.forward(x), axis=1)
    
    def plot_training_curves(self):
        
        plt.figure(figsize=(12, 4))
        plt.subplot(1, 2, 1)
        plt.plot(self.losses, label='Training Loss')
        plt.title('Training Loss Curve')
        plt.xlabel('Epochs')
        plt.ylabel('Loss')
        plt.legend()

        
        plt.subplot(1, 2, 2)
        plt.plot(self.val_accuracies, label='Validation Accuracy')
        plt.title('Validation Accuracy Curve')
        plt.xlabel('Epochs')
        plt.ylabel('Accuracy')
        plt.legend()
        
        plt.tight_layout()
        plt.show()

    
x_train =mnist_tr
y_train = one_hot_encoding(mnist_train_labels, 10)

x_test=mnist_ts
y_test=one_hot_encoding(mnist_test_labels, 10)

init_methods = ['zero', 'uniform_random']
results = {}
for init_method in init_methods:
    print(f"\nTraining with {init_method} initialization...")
    model = MLP(input_size=784, output_size=10, hidden_layers=[300, 200], init_method=init_method)
    model.fit(x_train, y_train, learning_rate=0.01, epochs=200, x_val=x_test, y_val=y_test)
    accuracy = model.evaluate(x_test, y_test)
    test_error = 1 - accuracy
    results[init_method] = {
        "test_error": test_error
    }

for init_method, result in results.items():
    print(f"{init_method} initialization Test Error: {result['test_error']:.4f}")
```
